## Data Visualization using Histogram

```python
df.plot.hist()
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.set_style('whitegrid')
plt.hist(df['sepal.length'],color='brown')
plt.show()
```

```python
df.hist(figsize=(9,6),color='skyblue')
plt.show()
```

```python
sns.histplot(data=df,x='sepal.length',hue='variety')
```

```python
plt.figure(figsize=(8,6))
sns.set_style('whitegrid')
plt.hist([df['sepal.length'],df['sepal.width'],df['petal.length'],df['petal.width']],label=['sepal.length','sepal.width','petal.length','petal.width'])
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.set_style('whitegrid')
plt.hist(df['sepal.length'],color='brown')
plt.hist(df['sepal.width'],color='blue',alpha=1)
plt.hist(df['petal.length'],color='green',alpha=0.6)
plt.hist(df['petal.width'],color='yellow',alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(20,13))
plt.subplot(2,2,1)
plt.hist(df['sepal.length'],20,density=True,histtype='bar',facecolor='b',alpha=0.5)

plt.subplot(2,2,2)
plt.hist(df['sepal.width'],color='indigo',alpha=1)

plt.subplot(2,2,3)
plt.hist(df['petal.length'],color='green',alpha=0.6)

plt.subplot(2,2,4)
plt.hist(df['petal.width'],color='magenta',alpha=0.7)

plt.show()
```

```python
sns.pairplot(df)
plt.show()
```

## Conclusion

The given iris dataset is first analysed properly,removing duplicates along with null values to make sure the data is fully cleaned. After data cleaning, histogram is being used to visualize the distribution of data in the given dataset, plotting sepal length & width and petal length & width to get a clear understanding of distributed data using different bin sizes and alpha values.
